```python
import mglearn
import matplotlib.pyplot as plt


from mglearn.datasets import make_forge
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

x, y = make_forge()
print(x.shape, y.shape)

count = 0
for label in y:
    if label == 0:
        count +=1
        
print(f"\n\nThe number of label 0 is {count}")
print(f"The number of label 1 is {len(y) - count}")

plt.ylim(0,20)
plt.bar(x = ["0", "1"], height = [count, len(y)-count], color = ['red', 'blue'])
plt.show()
plt.close()

x1 = x[:,0]
x2 = x[:,1]


mglearn.discrete_scatter(x1, x2, y)
plt.xlabel("Fist feature")
plt.ylabel("Second feature")
plt.legend(["Class 0", "Class 1"], loc = 4)
plt.show()
plt.close()


mglearn.plots.plot_knn_classification(n_neighbors=3)
plt.show()
plt.close()


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
knn = KNeighborsClassifier(3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
acc = (y_test == y_pred).sum()/len(y_test)*100
print("The model accuracy on the test set is {:2.2f}%".format(acc))

fig, axes = plt.subplots(1, 3, figsize=(10,5))

for n_neighbors, ax in zip([1,3,9], axes):
    knn = KNeighborsClassifier(n_neighbors).fit(x_train, y_train)
    mglearn.plots.plot_2d_separator(knn, x, ax = ax,
                            fill=True, eps=0.5)
    mglearn.discrete_scatter(x1, x2, y, ax=ax)
    ax.set_title(f'{n_neighbors} neighbor(s)')
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
    
axes[0].legend()
plt.show()
plt.close()
```
